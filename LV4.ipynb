```python
import pandas as pd
from sklearn.model_selection import train_test_split
import re
from keras.preprocessing.text import Tokenizer

data = pd.read_csv('SpamDetectionData.txt')


print(data.columns)

x = data['Message']
y = data['Label']

data['Message'] = data['Message'].apply(lambda x: re.sub(r'<p\s*\/?>', '', x))

print(data.head(10))

def count_letters(sentence):
    words = re.split(r'[.!?]', sentence)
    if len(words) >= 2:
      first_sentence = words[0].strip()
      second_sentence = words[1].strip()
      return len(first_sentence), len(second_sentence)
    else:
      return 0, 0

data['First_Sentence_Letters'], data['Second_Sentence_Letters'] = zip(*data['Message'].map(count_letters))


print("Broj slova u prvoj recenici: ", data['First_Sentence_Letters'].sum())
print("Broj slova u drugoj recenici: ", data['Second_Sentence_Letters'].sum())

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)

print("Dimenzije trenirajuceg skupa: ")
print("x_train: ", x_train.shape)
print("y_train: ", y_train.shape)
print()
print("dimenzije testnog skupa: ")
print("x_test: ", x_test.shape)
print("y_test: ", y_test.shape)

tokenizer = Tokenizer()
tokenizer.fit_on_texts(x_train)
train_data_seq = tokenizer.texts_to_sequences(x_train)
test_data_seq = tokenizer.texts_to_sequences(x_test)
print(tokenizer.word_index)
print(tokenizer.word_counts)




```
